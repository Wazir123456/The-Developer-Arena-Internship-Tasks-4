 DATA CLEANING & VALIDATION

```python


import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_csv("sales_data.csv")
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: sales_data.csv not found")
    exit()

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Remove rows with missing values
df.dropna(inplace=True)

# Validate numeric columns
numeric_cols = ['Quantity', 'Price', 'Total_Sales']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dropna(inplace=True)

# Create Month column
df['Month'] = df['Date'].dt.to_period('M')

print("\nData cleaning completed.")
print(df.head())



# Total sales by product
sales_by_product = df.groupby('Product')['Total_Sales'].sum()

# Monthly sales trend
monthly_sales = df.groupby('Month')['Total_Sales'].sum()

# Sales distribution by region
sales_by_region = df.groupby('Region')['Total_Sales'].sum()

print("\nSales by Product:")
print(sales_by_product)

print("\nMonthly Sales:")
print(monthly_sales)

print("\nSales by Region:")
print(sales_by_region)

```

DATA VISUALIZATION

```python

plt.figure()
sales_by_product.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()

plt.figure()
monthly_sales.sort_index().plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()


plt.figure()
sales_by_region.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution by Region")
plt.ylabel("")
plt.tight_layout()
plt.show()

```

INSIGHTS REPORT

```python


top_product = sales_by_product.idxmax()
top_region = sales_by_region.idxmax()
best_month = monthly_sales.idxmax()

print("\n--- BUSINESS INSIGHTS ---")
print(f"Top-selling product: {top_product}")
print(f"Highest sales region: {top_region}")
print(f"Best performing month: {best_month}")

print("\nProject completed successfully!")
```
